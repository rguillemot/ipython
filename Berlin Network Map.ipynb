```python
import pickle
output = open('all_stations.pkl', 'rb')
all_stations = pickle.load(output)
output.close()
output = open('stations_pos.pkl', 'rb')
stations_pos = pickle.load(output)
output.close()
output = open('shortpathdict.pkl', 'rb')
shortpathdict = pickle.load(output)
output.close()
```

```python
# -*- coding: utf-8 -*- 

import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rcParams
rcParams['text.latex.unicode']=True

groupe = [u'Birkenstra\xdfe',u'Sch\xf6nhauser Allee',u'Gneisenaustra\xdfe',u'Sch\xf6nleinstra\xdfe']

def distance_groupe_station(station,groupe):
    return sum([len(shortpathdict[(station,s)]) for s in groupe])

distances = {s:distance_groupe_station(s,groupe) for s in stations_pos.keys()}

min_dist = min(distances.values())
max_dist = max(distances.values())

cmap = plt.get_cmap('autumn')

f, ax = plt.subplots(figsize=(12,6))
    
for k in stations_pos.keys():
    x = stations_pos[k][1]
    y = stations_pos[k][0]
    idx = min(distances[k]-min_dist,4)
    ax.scatter(x,y,marker='o',c=cmap(idx*255/4),label=k,s=250/(idx+1))
    
x_ring = []
#for k in all_stations['s42']:
#    ax.plot(stations_pos[k][1],stations_pos[k][0],marker='o',c='red')

for k in groupe:
    ax.scatter(stations_pos[k][1],stations_pos[k][0],s=300, facecolors='none', edgecolors='black')


ax.set_xlim(13.27, 13.49)
ax.set_ylim(52.46, 52.56)
```
